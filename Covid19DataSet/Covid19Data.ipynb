```python
coviddf['State'].unique()
```

As there are null values we can replce those null values with some comment as 'Not Defined'

```python
coviddf['State'] = coviddf['State'].fillna('Not Defined')
```

```python
coviddf.head()
```

We want to get the data region wise then we can use Groupby function for it

```python
# group by object
region_grp = coviddf.groupby('Region')
```

```python
# we can sum up the data to get the count for different regions
region_grp.sum().head(20)
```

```python
# we can get get the top 20 coutries till date but sorting the values on the basis of Confirmned cases
# we can use the below syntax for single column to display and by default sorted by that column
region_grp['Confirmed'].sum().sort_values(ascending = False).head(20)
```

```python
# display confirmed and deaths sorted by confirmed column
region_grp['Confirmed', 'Deaths'].sum().sort_values(ascending = False, by=['Confirmed']).head(20)
```

```python
# we can use the below syntax to display all columns and sort by a single cloumn
region_grp.sum().sort_values(ascending = False, by=['Confirmed']).head(20)
```

```python
# we can sort the regions on basis of two columns

region_grp.sum().sort_values(ascending = False, by=['Confirmed', 'Deaths']).head(20)
```
